```python
%%time
dfEstudo=dfEstudo.sort_values(by=["DESC_ARTIGO", "STORE", "DATA"])
```

# Tempo para a reposição

```python
dfEstudo["movimento"] = np.where((dfEstudo["EXPECTED"] > 0) | (dfEstudo["INTRANSIT"] > 0), 1, 0)
```

- Somar os dias para 

```python
groups = (dfEstudo['movimento'] != dfEstudo['movimento'].shift()).cumsum()
result = dfEstudo.groupby(groups).agg({'DATA': 'first', 'DESC_ARTIGO': 'first', 'STORE': 'first', 'movimento': 'sum'}).reset_index(drop=True)
result = result[result['movimento'] > 0]
```

```python
tempo_mean = result['movimento'].mean()
tempo_min = result['movimento'].min()
tempo_max = result['movimento'].max()
```

```python
print("Tempo médio para uma reposição:", tempo_mean+1, "dias.")
print("Tempo mínimo de reposição:", tempo_min+1, "dias.")
print("Tempo máximo de reposição:", tempo_max+1, "dias.")
```
