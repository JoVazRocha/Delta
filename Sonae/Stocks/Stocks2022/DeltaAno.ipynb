```python
%%time
import pandas as pd
import os
import glob
import pathlib2 as pl2
```

# Ler os ficheiros na pasta

- Demora 15 min

```python
%%time

a=0
# Define the directory containing the Excel files
directory = 'D:\\Brands and Ninjas\\Dadosjanout'

# Define an empty list to store the DataFrames
dataframes = []

# Iterate through the Excel files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".xlsx") and filename.startswith("Info_DELTA_"):
        a+=1
        print (a)
        # Read the data from the Excel file into a DataFrame
        filepath = os.path.join(directory, filename)
        df = pd.read_excel(filepath)
        
        # Add the DataFrame to the list
        dataframes.append(df)

# Concatenate all the DataFrames into a single DataFrame
original = pd.concat(dataframes, ignore_index=True)
```

# Fazer datetime e Seleccionar colunas
- 15 seg

```python
%%time
dfTeste=original.loc[:, ['DATA', 'EAN', 'DESC_ARTIGO', 'STORE', "STORE_NAME", 
                         "SOH", "INTRANSIT", "EXPECTED", "PRES_STOCK", "VND (D-1)"]].copy()

dfTeste['DATA'] = pd.to_datetime(original['DATA'], format='%d-%m-%Y')
```

> Imaginando que há falhas a nível de anos (há um 2021 e não deveria haver)

```python
%%time
dfTeste.loc[dfTeste['DATA'].dt.year == 2021, 'DATA'] = dfTeste.loc[dfTeste['DATA'].dt.year == 2021, 'DATA'] + pd.offsets.DateOffset(years=1)
```
